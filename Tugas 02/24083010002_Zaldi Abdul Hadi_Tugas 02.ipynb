```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# interval
x = np.linspace(0, 10, 500)

# deret fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
h = 0.5
```

```python
# 1. Hitung diferensial dari fungsi ‘f’ untuk interval x
```

```python
f_fwd = (f[1:] - f[:-1]) / h
f_bwd = (f[1:] - f[:-1]) / h
f_cen = (f[2:] - f[:-2]) / (2*h)

x_fwd = x[:-1]
x_bwd = x[1:]
x_cen = x[1:-1]

print("Panjang Forward Difference :", len(f_fwd))
print("Panjang Backward Difference:", len(f_bwd))
print("Panjang Central Difference :", len(f_cen))

plt.figure(figsize=(8,4))
plt.plot(x, f, 'b-', label="f(x)", linewidth=2)
plt.plot(x_fwd, f_fwd, 'g--', label="Forward difference", linewidth=1.5)
plt.plot(x_bwd, f_bwd, 'r:', label="Backward difference", linewidth=1.5)
plt.plot(x_cen, f_cen, 'm-.', label="Central difference", linewidth=1.5)
plt.title("Diferensial f(x)")
plt.legend(); plt.grid(True)
plt.show()
```

Dari grafik terlihat bahwa metode forward dan backward difference cenderung memiliki sedikit deviasi terhadap perubahan fungsi, sedangkan central difference memberikan hasil yang lebih halus dan stabil mengikuti pola f(x). Hal ini menunjukkan bahwa central difference umumnya lebih akurat dalam menghampiri nilai turunan sebenarnya dibandingkan dengan metode forward maupun backward.

```python
# 2. Hitung integral dari fungsi ‘f’ untuk interval x
```

```python
F_total = np.cumsum((f[:-1] + f[1:]) / 2 * h)   
F_total = np.insert(F_total, 0, 0)

print("Nilai Integral Total   :", F_total[-1])

plt.figure(figsize=(8,4))
plt.plot(x, F_total, 'k', label="Integral Total")
plt.title("Integral f(x)")
plt.legend(); plt.grid(True)
plt.show()
```

Grafik ini menunjukkan integral dari sebuah fungsi f(x) dengan nilai integral total 54.55, grafik ini secara sempurna menggambarkan akumulasi (bisa positif atau negatif) dari suatu nilai dari waktu ke waktu.

```python
# 3. Hitung secara terpisah bagian integral di atas dan di bawah y=0
```

```python
# Memisahkan bagian atas dan bawah seperti konsep "y_upper" dan "y_lower"
f_upper = np.clip(f, 0, None)
f_lower = np.clip(f, None, 0)

# Menghitung integral menggunakan metode trapezoidal rule (mirip dengan menghitung luas lingkaran)
area_upper = np.trapz(f_upper, x)
area_lower = np.trapz(f_lower, x)

print("Luas Integral di Atas Sumbu x  :", area_upper)
print("Luas Integral di Bawah Sumbu x :", area_lower)

plt.figure(figsize=(8,5))
plt.plot(x, f, 'k', label='f(x)')
plt.fill_between(x, f_upper, 0, alpha=0.5, label='Area atas')
plt.fill_between(x, f_lower, 0, alpha=0.5, label='Area bawah')
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Integral f(x) di Atas & Bawah Sumbu x")
plt.legend()
plt.grid(True)
plt.show()
```

Grafik ini menampilkan fungsi f(x) serta memvisualisasikan kontribusi integralnya secara terpisah. Area biru di atas sumbu x menunjukkan total semua nilai positif dari fungsi, yang jika diintegralkan berjumlah 4.60, sedangkan area oranye di bawah sumbu x menunjukkan total nilai negatifnya, yang berjumlah -2.42. Dengan demikian, nilai integral total atau "area bersih" dari fungsi ini pada interval yang ditampilkan adalah hasil penjumlahan keduanya, yaitu sekitar 2.186.
