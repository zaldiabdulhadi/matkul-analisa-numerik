```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of points
n = 200
NPM = 2
```

```python
# Generate random x values between 0 and 4Ï€
np.random.seed(NPM)
x = np.random.uniform(0, 4*np.pi, n)
```

```python
# Original function
y_true = 4 * np.sin(2*x + 0.2)
```

```python
# Add Gaussian noise
noise = np.random.normal(0, 0.4, n)  # mean=0, std=0.5
y_noisy = y_true + noise
```

```python
# Plot
plt.scatter(x, y_noisy, label="Noisy data", color="red", alpha=0.6)
# plt.plot(np.sort(x), 4*np.sin(4*np.sort(x) + 0.2), label="True function", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# TUGAS: temukan kembali fungsi asli dengan interpolasi dari titik-titik data acak (y_noisy)
```

```python
orders = [3, 5, 9, 12, 15] 
x_grid = np.linspace(0, 4*np.pi, 500)

plt.figure(figsize=(10,5))
plt.scatter(x, y_noisy, label="Noisy data", color="red", alpha=0.4)

for k in orders:
    coeffs = np.polyfit(x, y_noisy, k)     # cari koefisien polinomial orde k
    y_poly = np.polyval(coeffs, x_grid)    # evaluasi polinomial pada grid
    plt.plot(x_grid, y_poly, label=f"Polinom orde {k}")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Interpolasi Polinomial dari Titik Data Acak")
plt.show()
```

Grafik menunjukkan bahwa polinomial orde rendah (3 dan 5) tidak mampu mengikuti pola sinus sama sekali, sedangkan orde 9 mulai mendekati namun masih kurang akurat. Polinomial orde 12 dan terutama orde 15 jauh lebih mendekati bentuk sinus asli, mengikuti puncak dan lembah data dengan baik. Dengan demikian, orde 15 adalah yang paling mendekati fungsi asal dan menjadi pilihan terbaik untuk merekonstruksi bentuk sinus dari data noisy.
